```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies=movies[movies['overview'].notna()]
```

```python
movies.isna().sum()
```

```python
movies.columns
```

```python
movies=movies.drop(['budget', 'genres', 'homepage', 'id', 'keywords', 'original_language','production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime',
       'spoken_languages', 'status', 'tagline', 'title'],axis=1)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(min_df=3,ngram_range=(1,3),stop_words='english',analyzer='word')
tf_matrix=tfidf.fit_transform(movies['overview'])
print(tf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sgm=sigmoid_kernel(tf_matrix,tf_matrix)
```

```python
sgm
```

```python
indices=pd.Series(movies.index,index=movies['original_title']).drop_duplicates()
```

```python
indices
```

```python
def recommend(name,sgm=sgm):
    idx=indices[name]
    movie_scores=list(enumerate(sgm[idx]))
    movie_scores=sorted(movie_scores,key=lambda x:x[1],reverse=True)
    top_10_movies=movie_scores[1:11]
    movies_indices=[i[0] for i in top_10_movies]
    return movies['original_title'].iloc[movies_indices]
```

```python
recommend('John Carter')
```

```python
recommend('The Dark Knight Rises')
```
